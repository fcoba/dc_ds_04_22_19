```python
### 3. Load the data into pandas df
df = pd.read_csv('master.csv')
df
```

```python
### 4. Pick only rows where country starts with U `use apply and lambda`
df[df['country'].apply(lambda x: x.startswith('U'))]
```

```python
### 5. Get the names of unique countries left `use pd.unique` 
print(df[~df['country'].apply(lambda x: x.startswith('U'))]['country'].unique())
```

```python
### 6. Group by age column and sum # suicides
df_us.groupby(['age']).sum()
```

```python
### 7. Change the age column to a categorical one 
# `use astype`

df_us['age'] = df_us['age'].astype('category')
```
